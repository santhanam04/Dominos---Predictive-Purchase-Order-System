```python
import pandas as pd

# Load the CSV files
df1 = pd.read_csv('Pizza_Sale - pizza_sales.csv')
df2 = pd.read_csv('Pizza_ingredients - Pizza_ingredients.csv')

# Merge on 'pizza_name_id'
merged_df = pd.merge(df1, df2, on='pizza_name_id', how='inner')

# View result
merged_df
```

```python
for col in merged_df.select_dtypes(include=['float64', 'int64']):
    if merged_df[col].isna().sum() > 0:
        median_val = merged_df[col].median()
        print(f"Filling NaN in '{col}' with median: {median_val}")
        merged_df[col] = merged_df[col].fillna(median_val)
```

```python
for col in merged_df.select_dtypes(include=['object', 'category']):
    if merged_df[col].isna().sum() > 0:
        mode_val = merged_df[col].mode()[0]
        print(f"Filling NaN in '{col}' with mode: {mode_val}")
        merged_df[col] = merged_df[col].fillna(mode_val)
```

```python
# Function to remove outliers from numerical columns
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]


```

```python
# Apply to numerical columns
for col in merged_df.select_dtypes(include=['float64', 'int64']).columns:
    merged_df = remove_outliers_iqr(merged_df, col)
```

```python
merged_df['order_date'] = merged_df['order_date'].str.replace('-', '/', regex=False)
merged_df['order_date'] = pd.to_datetime(merged_df['order_date'], dayfirst=True, errors='coerce')
```

```python
merged_df['order_year'] = merged_df['order_date'].dt.year
merged_df['order_month'] = merged_df['order_date'].dt.month
merged_df['order_day'] = merged_df['order_date'].dt.day
merged_df['day_of_week'] = merged_df['order_date'].dt.day_name()
merged_df['week'] = merged_df['order_date'].dt.isocalendar().week
```

```python
# Step 1: Calculate total sales per date
daily_sales = merged_df.groupby('order_date')['total_price'].sum()

# Step 2: Map daily total sales back to the original DataFrame
merged_df['daily_total_sales'] = merged_df['order_date'].map(daily_sales)
```

```python
monthly_sales = merged_df.groupby(['order_year', 'order_month'])['total_price'].sum().reset_index()

# Rename columns to match what pd.to_datetime expects
monthly_sales = monthly_sales.rename(columns={'order_year': 'year', 'order_month': 'month'})

# Add a day column for complete date
monthly_sales['day'] = 1

# Now safely convert to datetime
monthly_sales['Year-Month'] = pd.to_datetime(monthly_sales[['year', 'month', 'day']])
```

```python
# Convert to datetime (temporary) for extraction
order_time_dt = pd.to_datetime(merged_df['order_time'].astype(str), format='%H:%M:%S', errors='coerce')

# Extract parts
merged_df['order_hour'] = order_time_dt.dt.hour
merged_df['order_minute'] = order_time_dt.dt.minute
merged_df['order_second'] = order_time_dt.dt.second
```

```python
from pandas.tseries.holiday import USFederalHolidayCalendar

calendar = USFederalHolidayCalendar()
holidays = calendar.holidays(start=merged_df['order_date'].min(), end=merged_df['order_date'].max())

merged_df['is_holiday'] = merged_df['order_date'].isin(holidays).astype(int)
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

merged_df['season'] = merged_df['order_month'].apply(get_season)
```

```python
merged_df['is_promo_month'] = merged_df['order_month'].isin([11, 12]).astype(int)
```

```python
merged_df.drop(['order_time'], axis=1, inplace=True)
```

```python
merged_df['pizza_name'] = merged_df['pizza_name_x']
merged_df.drop(['pizza_name_x', 'pizza_name_y'], axis=1, inplace=True)
```

```python
def merge_ingredients(row):
    if pd.isna(row['pizza_ingredients_x']):
        return row['pizza_ingredients_y']
    if pd.isna(row['pizza_ingredients_y']):
        return row['pizza_ingredients_x']
    
    # Combine and deduplicate ingredients
    ingredients_x = [ing.strip() for ing in row['pizza_ingredients_x'].split(',')]
    ingredients_y = [ing.strip() for ing in row['pizza_ingredients_y'].split(',')]
    
    combined = list(dict.fromkeys(ingredients_x + ingredients_y))  # preserve order, remove duplicates
    return ', '.join(combined)

# Apply function to rows
merged_df['pizza_ingredients'] = merged_df.apply(merge_ingredients, axis=1)

# Drop old columns
merged_df.drop(['pizza_ingredients_x', 'pizza_ingredients_y'], axis=1, inplace=True)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_percentage_error

# ✅ Step 1: Prepare your DataFrame
# Replace this with your actual DataFrame
# Example:
# daily_sales = pd.read_csv("your_sales_data.csv")  # with columns 'ds' and 'y'

daily_sales['ds'] = pd.to_datetime(daily_sales['ds'])
daily_sales_arima = daily_sales.set_index('ds')
daily_sales_arima = daily_sales_arima.asfreq('D')  # Daily frequency

# Fill any missing values (optional, depending on your data)
daily_sales_arima['y'] = daily_sales_arima['y'].interpolate()

# ✅ Step 2: Hold out last 7 days for evaluation
y_true = daily_sales_arima['y'][-7:].values
train_data = daily_sales_arima['y'][:-7]

# ✅ Step 3: Fit ARIMA model
model_arima = ARIMA(train_data, order=(5, 1, 0))  # (p,d,q)
arima_result = model_arima.fit()

# ✅ Step 4: Forecast next 7 days
forecast_arima = arima_result.forecast(steps=7)

# ✅ Step 5: Evaluate and Plot
print("Next 7-day forecast:")
print(forecast_arima)

# Evaluate with MAPE
mape_arima = mean_absolute_percentage_error(y_true, forecast_arima) 
print(f"ARIMA MAPE: {mape_arima:.4f}")

# ✅ Optional: Plot results
plt.figure(figsize=(10, 4))
plt.plot(daily_sales_arima.index, daily_sales_arima['y'], label='Historical Sales')
plt.plot(forecast_arima.index, forecast_arima.values, label='Forecast', color='red')
plt.axvline(daily_sales_arima.index[-7], color='gray', linestyle='--', label='Forecast Start')
plt.title('ARIMA Forecast - Next 7 Days')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

# === STEP 1: Prepare and scale the data ===
# Ensure 'daily_sales' has 'ds' (date) and 'y' (sales)
# Example: daily_sales = pd.read_csv("your_file.csv")
# daily_sales['ds'] = pd.to_datetime(daily_sales['ds'])

scaler = MinMaxScaler()
scaled_sales = scaler.fit_transform(daily_sales['y'].values.reshape(-1, 1))

# === STEP 2: Create LSTM sequences ===
def create_sequences(data, seq_length=7):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

X_lstm, y_lstm = create_sequences(scaled_sales, seq_length=7)
X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

# === STEP 3: Build the LSTM model ===
model_lstm = Sequential([
    Input(shape=(7, 1)),
    LSTM(50, activation='relu'),
    Dense(1)
])
model_lstm.compile(optimizer='adam', loss='mse')

# === STEP 4: Train the LSTM model ===
model_lstm.fit(X_lstm, y_lstm, epochs=20, verbose=1)

# === STEP 5: Forecast next 7 days ===
last_seq = scaled_sales[-7:]  # Last 7 days for input
future_preds = []

for _ in range(7):
    input_seq = last_seq.reshape(1, 7, 1)
    pred = model_lstm.predict(input_seq, verbose=0)[0][0]
    future_preds.append(pred)
    last_seq = np.append(last_seq[1:], [[pred]], axis=0)  # Slide window

# === STEP 6: Inverse transform the predictions ===
future_preds = scaler.inverse_transform(np.array(future_preds).reshape(-1, 1)).flatten()

# === STEP 7: Display forecast ===
print("Next 7-day LSTM forecast:")
print(future_preds)

# === STEP 8: Evaluate with MAPE (optional) ===
y_true = daily_sales['y'].values[-7:]
mape = mean_absolute_percentage_error(y_true, future_preds) 
print(f"LSTM MAPE: {mape:.4f}")

# === Optional: Plot forecast ===
plt.figure(figsize=(10, 4))
plt.plot(daily_sales['ds'], daily_sales['y'], label='Historical Sales')
future_dates = pd.date_range(daily_sales['ds'].iloc[-1] + pd.Timedelta(days=1), periods=7)
plt.plot(future_dates, future_preds, label='LSTM Forecast', color='red', marker='o')
plt.title('LSTM Forecast - Next 7 Days')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error

# === STEP 1: Generate lag features ===
def create_lag_features(series, lags=7):
    df = pd.DataFrame({'y': series})
    for i in range(1, lags + 1):
        df[f'lag_{i}'] = df['y'].shift(i)
    return df.dropna()

# === STEP 2: Prepare the data ===
# Assume daily_sales is already loaded with 'y' column
lags = 7
df_lagged = create_lag_features(daily_sales['y'], lags=lags)

X = df_lagged[[f'lag_{i}' for i in range(1, lags + 1)]]
y = df_lagged['y']

# === STEP 3: Train Linear Regression model ===
model_lr = LinearRegression()
model_lr.fit(X, y)

# === STEP 4: Forecast next 7 days ===
last_known = daily_sales['y'].values[-lags:].tolist()
future_preds_lr = []

for _ in range(7):
    input_seq = pd.DataFrame([last_known[-lags:]], columns=[f'lag_{i}' for i in range(1, lags + 1)])
    pred = model_lr.predict(input_seq)[0]
    future_preds_lr.append(pred)
    last_known.append(pred)

# === STEP 5: Evaluate with MAPE (optional, if true values available) ===
if len(daily_sales) >= lags + 7:
    y_true = daily_sales['y'].values[-7:]
    mape_lr = mean_absolute_percentage_error(y_true, future_preds_lr)
    print(f"\nLinear Regression MAPE: {mape_lr:.4f}")

# === STEP 6: Display Forecast ===
print("\nNext 7-day forecast (Linear Regression):")
for i, val in enumerate(future_preds_lr, 1):
    print(f"Day {i}: {val:.2f}")
```

```python
import pandas as pd
from collections import defaultdict
import ast

# Step 1: Clean 'pizza_ingredients'
merged_df['pizza_ingredients'] = merged_df['pizza_ingredients'].astype(str).str.lower().str.strip()

# Step 2: Ensure 'Items_Qty_In_Grams' is always a list of floats
def parse_quantities(x):
    if isinstance(x, list):
        return x
    try:
        x = ast.literal_eval(str(x))
        if isinstance(x, (int, float)):
            return [float(x)]
        return list(map(float, x)) if isinstance(x, (list, tuple)) else [float(x)]
    except:
        return [0.0]  # fallback for invalid rows

merged_df['Items_Qty_In_Grams'] = merged_df['Items_Qty_In_Grams'].apply(parse_quantities)

# Step 3: Initialize dictionary
ingredient_totals = defaultdict(float)

# Step 4: Calculate totals
for _, row in merged_df.iterrows():
    ingredients = [i.strip() for i in row['pizza_ingredients'].split(',')]
    quantities = row['Items_Qty_In_Grams']
    qty_sold = int(row['quantity'])

    # Zip ingredients with their quantities
    for ingredient, qty_per_pizza in zip(ingredients, quantities):
        ingredient_totals[ingredient] += qty_per_pizza * qty_sold

# Step 5: Output to DataFrame
ingredient_summary = pd.DataFrame([
    {'ingredient': ing, 'total_quantity_required_in_grams': qty}
    for ing, qty in ingredient_totals.items()
]).sort_values(by='ingredient').reset_index(drop=True)

# Step 6: Display
print(ingredient_summary)
```
